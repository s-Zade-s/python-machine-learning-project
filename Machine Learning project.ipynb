# CAN WE LOCATE BEST VALUE FOR K TO GET BEST ACCURACY?¶

```python
k_range=range(1,100)
scores=[]
for k in k_range:
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train,y_train)
    y_pred=knn.predict(x_test)
    scores.append(metrics.accuracy_score(y_test,y_pred))
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(k_range,scores)
plt.xlabel('Range of values of K')
plt.ylabel('Accuracy score')
plt.show()
```

```python
from sklearn.cross_validation import cross_val_score
knn=KNeighborsClassifier(n_neighbors=5)
scores=cross_val_score(knn,X,y,cv=10,scoring='accuracy')
print(scores)
print(scores.mean())
```

```python
k_range= range(1,31)
k_scores=[]
for k in k_range:
    knn=KNeighborsClassifier(n_neighbors=k)
    scores=cross_val_score(knn,X,y,cv=10,scoring='accuracy')
    k_scores.append(scores.mean())
print(k_scores)
```

```python
plt.plot(k_range,k_scores)
plt.xlabel('Value of K for KNN')
plt.ylabel('Cross-validated accuracy')
plt.title('CV = 10')
plt.show()
```

# Tuning parameters more efficiently on GridSearchCV

```python
from sklearn.grid_search import GridSearchCV
# This is an arbitrary range for K-values
k_range=np.arange(1,31)
print(k_range)
```

```python
param_grid=dict(n_neighbors=k_range)
print(param_grid)
```

```python
grid=GridSearchCV(knn,param_grid,cv=10,scoring='accuracy')
```

```python
#now, to fit the grid with the data
grid.fit(X,y)
```

```python
grid.grid_scores_
```

```python
print(grid.grid_scores_[0].parameters)
print(grid.grid_scores_[0].cv_validation_scores)
print(grid.grid_scores_[0].mean_validation_score)
```

```python
grid_mean_scores=[result.mean_validation_score for result in grid.grid_scores_]
print(grid_mean_scores)
```

```python
plt.plot(k_range,grid_mean_scores)
plt.xlabel('Values of K for KNN')
plt.ylabel('Cross-Validation accuracy')
plt.show()
```

```python
#and, voila, here's the final result
print(grid.best_score_)
print(grid.best_params_)
print(grid.best_estimator_)
```

# Searching multiple parameters simultaneously

# Reducing computational expense using RandomisedSearchCV

```python
#So, some of the tasks above were extremely computationally expensive and I have learned that it is just...
#...not feasible to search many different parameters. I therefore experimented with, and learned, the RandomizedSearchCV method
```

```python
from sklearn.grid_search import RandomizedSearchCV
```

```python
#specify the 'parameter distributions' rather than a 'parameter grid'
param_dist=dict(n_neighbors=k_range,weights=weight_options)
```

```python
#n_iter parameter controls the number of searches
rand=RandomizedSearchCV(knn,param_dist,cv=10,scoring='accuracy',n_iter=4,random_state=5)
rand.fit(X,y)
rand.grid_scores_
```

```python
print(rand.best_score_)
print(rand.best_params_)
```
